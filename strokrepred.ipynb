# Stroke Prediction

```python
from strokepred.dataset import create_dataset
import matplotlib.pyplot as plt
import pandas as pd

df = create_dataset()
print(df)
```

```python
# Print gender distribution chart
df.groupby("gender").count().plot.pie(y="stroke", autopct='%1.1f%%')
```

```python
# Print age distribution chart
df["age"].plot.hist()
```

```python
# Heatmap of correlation between age and stroke
corr_matrix = pd.get_dummies(df).corr()
plt.matshow(corr_matrix)
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.colorbar()
```
